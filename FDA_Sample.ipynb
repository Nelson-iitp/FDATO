# FDA Sample Benchmark

```python
import fda
import numpy as np
import matplotlib.pyplot as plt
```

# Fitness Function

```python
# minimize this function
def fitness(x):
    return np.sum(x**2)
```

# FDA Agorithm

```python
F = fda.FDA(5, -100, 100, fitness, alpha=6, beta=4)
```

# FDA Soultions

```python
best_flow, best_flow_cost, Flow_X, Flow_fitness, W_var_history, C_history, V_history  = F.optimize(200, verbose=2)
```
